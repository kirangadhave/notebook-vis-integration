```python
source = data.cars()
```

```python
brush = alt.selection_interval(name="brs")

alt.Chart(source, name="sp").mark_point().encode(
    x='Miles_per_Gallon:Q',
    y='Horsepower:Q',
    color=alt.condition(brush, alt.value('steelblue'), alt.value('grey')),
).add_params(brush)
```
