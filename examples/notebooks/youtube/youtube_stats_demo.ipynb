```python
import pandas as pd
import altair as alt
import persist_ext as PR

PR.enable('altair')
```

```python
df = pd.read_csv("youtube_stats.csv")
df
```

```python
df = df.dropna()
df
```

```python
pt_sel = alt.selection_point(name="pt_sel", encodings=["x"])

alt.Chart(df, name="bars").mark_bar().encode(
    x="category:N",
    y="count()",
    color=alt.condition(pt_sel, alt.ColorValue("steelblue"), alt.ColorValue("grey"))
).add_params(pt_sel).properties(width=600)
```
