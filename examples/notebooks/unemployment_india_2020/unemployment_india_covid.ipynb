```python
df = pd.read_csv("./unemployment_india_2020.csv")
df
```

```python
selector = alt.selection_point(name="selector", encodings=["x"])

alt.Chart(df, name="bars").mark_bar().encode(x="Region:N",
            y="count()",
            color=alt.condition(selector, alt.value('steelblue'), alt.value('grey'))
           ).add_params(selector)
```

```python
cleaned_df[" Date"] = pd.to_datetime(cleaned_df[" Date"]).dt.strftime("%B")
cleaned_df.head()
```

```python
selector = alt.selection_point(name="selector", encodings=["x"])

alt.Chart(cleaned_df, name="bars").mark_bar().encode(
            x=alt.X(" Date:N").sort("x"),
            y="mean( Estimated Unemployment Rate (%)):Q",
            color=alt.condition(selector, alt.value('steelblue'), alt.value('grey'))
           ).add_params(selector)
```
